```python
x1=data["Time Spent on Site"].mean()
x1
```

```python
x2=data["Page Views"].mean()
x2
```

```python
x3=data["Device Type"].mean()
x3
```

```python
x4=data["Previous Purchases"].mean()
x4
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
data["Made Purchase"] = label_encoder.fit_transform(data["Made Purchase"])
data["Made Purchase"]
```

```python
x5=data["Made Purchase"].mode()
x5
```

```python
data["Time Spent on Site"].fillna(x1, inplace = True)
data["Page Views"].fillna(x2, inplace = True)
data["Device Type"].fillna(x3, inplace = True)
data["Previous Purchases"].fillna(x4, inplace = True)
data["Made Purchase"].fillna(x5, inplace = True)
```

```python
missingvalue = data.isnull().sum()
missingvalue
```

```python
data
```

```python
x=data.drop(['User ID','Made Purchase'],axis=1)
y=data["Made Purchase"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy = accuracy_score(y_test, y_pred) 
accuracy
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

# 2. Plot confusion matrix using seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True ,xticklabels=["No Purchase", "Purchased"],
            yticklabels=["No Purchase", "Purchased"])

# 3. Add labels
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.title('Confusion Matrix')
plt.show()
```
